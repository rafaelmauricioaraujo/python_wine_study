# EDA com visualizações
visualizações para resopnder as perguntas.

    1 Histogramas das colunas Acidez fixa, total de dióxido de enxofre, pH, álcool neste conjunto de dados
    2 Gráficos de dispersão de qualidade em relação a variáveis Acidez volátil, açúcar residual, pH, álcool

```python
# Carregando o conjunto de dados
import pandas as pd

wine_df = pd.read_csv('winequality_edited.csv')
wine_df.head()
```

### Histogramas para diversas características

```python
# gráfico com a distribuição da acidez fixa
wine_df["fixed acidity"].hist();
```

```python
# gráfico com a distribuição do total de dióxido de enxofre
wine_df["total sulfur dioxide"].hist()
```

```python
# gráfico com a distribuição do pH
wine_df["pH"].hist()
```

```python
# gráfico com a distribuição da alcool
wine_df["alcohol"].hist()
```

```python
# gráfico com a relação entre qualidade do vinho e acidez volátil
wine_df.plot(kind="scatter", x="quality", y="volatile acidity");
```

```python
# gráfico com a relação entre qualidade do vinho e açúcar residual
wine_df.plot(kind="scatter", x="quality", y="residual sugar");
```

```python
# gráfico com a relação entre qualidade do vinho e pH
wine_df.plot(kind="scatter", x="quality", y="pH")
```

```python
# gráfico com a relação entre qualidade do vinho e pH
wine_df.plot(kind="scatter", x="quality", y="alcohol")
```

```python
# gráfico com a matriz de dispersão da tabela
pd.plotting.scatter_matrix(wine_df, figsize=(20,20));
```
