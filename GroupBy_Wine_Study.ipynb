### Utilizando o groupby do pandas para obter informações preciosas

```python
import pandas as pd

wine_df = pd.read_csv('winequality_edited.csv')
wine_df.head()
```

```python
#valor médio de todos os dados por colunas
wine_df.mean()
```

```python
#utilizando groupby para verificar o nível médio de cada coluna para cada nível de qualidade
wine_df.groupby('quality').mean()
```

```python
#utilizando o groupby para formar uma lista
wine_df.groupby(['quality', 'color']).mean()
```

```python
#agrupando sem índices
wine_df.groupby(['quality', 'color'], as_index=False).mean()
```

```python
#sem 
wine_df.groupby(['quality', 'color'], as_index=False).mean()
```
